```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### How to get the raw URL of a file on GitHub:

1. Navigate to the file on GitHub (e.g., a `.csv` file).
2. Click on the 'Raw' button. This will open the raw file content in a new tab.
3. Copy the URL from this new tab. This is the direct link to the raw file content.

Once you have the raw URL, you can use libraries like `pandas` to read it directly.

```python
import pandas as pd

# Example: Replace this with the raw URL of your GitHub file
github_raw_url = 'https://raw.githubusercontent.com/rq-goh/ADALL_github/main/laptop_prices_2024_sgd_TL.csv'

try:
    df = pd.read_csv(github_raw_url)
    print("Successfully loaded data from GitHub!")
    display(df.head())
except Exception as e:
    print(f"Error loading data: {e}")
    print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")
```

```python
#plot each df col graph
for col in df.columns:
  plt.figure(figsize=(8, 6))
  sns.histplot(df[col], kde=True)
  plt.title(f'Distribution of {col}')
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.show()
```
